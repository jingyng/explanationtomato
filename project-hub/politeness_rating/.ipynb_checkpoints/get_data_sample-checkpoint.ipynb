```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_files/politeness_rewriting_data.csv')
```

```python
bad_users = ['6093c341c4c80556c807a38d','64141beda1cef49212be9b6c','63b6d9a59f04985573644798']
df = df[(df.len>=7)&(df.edit_distance>=5)&(~df.user.isin(bad_users))]
```

```python
df
```

```python
s_df = pd.DataFrame()
s_df['text_raw'] = pd.concat([df['revised email'],df['displayed_text']])
s_df['instance_id_raw'] = pd.concat([df['instance_id'],df['instance_id']])
s_df['user_raw'] = list(df['user']) + ['enron']*len(df)
s_df['text'] = [it.replace('\n','<br>').replace('\r','<br>') for it in s_df['text_raw']]
s_df['inferred_politeness'] = pd.concat([df['new_politeness'],df['old_politeness']])
s_df['source'] = ['revised']*len(df) + ['enron']*len(df)
s_df['len'] = [len(it.split(' ')) for it in s_df['text']]
```

```python
s_df = s_df.drop_duplicates('text')
s_df['id'] = [str(i) for i in range(len(s_df))]
```

```python
for p in [25,50,75,90]:
    print(np.percentile(s_df['len'],p))
```
